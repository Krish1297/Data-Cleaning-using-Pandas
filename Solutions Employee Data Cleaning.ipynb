```python
import numpy as np
import pandas as pd
```

```python
data_frame = pd.read_csv("Employee Data Cleaning.csv")
print(data_frame)
```

```python
data_frame_copy = data_frame.copy()
```

```python

missing_data = data_frame_copy.isnull().sum()
print(missing_data)

averageAge = data_frame_copy["Age"].mean()
averageRating = data_frame_copy["Rating"].mean()

averageAge
averageRating

data_frame_copy["Age"] = data_frame_copy.Age.fillna(averageAge)
data_frame_copy["Rating"] = data_frame_copy.Rating.fillna(averageRating)


data_frame_copy['Age'] = data_frame_copy['Age'].astype(int)
data_frame_copy["Rating"] = data_frame_copy.Rating.round(decimals=1)


data_frame_copy
```

```python
data_frame_copy['Established'] = data_frame_copy['Established'].replace(-1, np.nan)
median_established = data_frame_copy['Established'].median()
data_frame_copy['Established'].fillna(median_established, inplace=True)
data_frame_copy['Established'] = data_frame_copy['Established'].astype(int)

data_frame_copy
```

```python
data_frame_copy['Easy Apply'] = data_frame_copy['Easy Apply'].replace("-1", "FALSE")


data_frame_copy['Rating'] = data_frame_copy['Rating'].replace(-1, 0)




data_frame_copy
```

```python
data_frame_copy['Location'] = data_frame_copy['Location'].str.replace(",In", "")
data_frame_copy['Location'] = data_frame_copy['Location'].str.replace(",Ny", "")
data_frame_copy['Location'] = data_frame_copy['Location'].str.replace(" In", "")
data_frame_copy['Location'] = data_frame_copy['Location'].str.replace(" Aus", "")

data_frame_copy
```

```python
data_frame_copy['Salary'] = data_frame_copy['Salary'].str.replace("k", "000")
data_frame_copy['Salary'] = data_frame_copy['Salary'].str.replace("$", "")

data_frame_copy
```

```python
data_frame_copy.reset_index(drop=True, inplace=True)
data_frame_copy.drop(columns=['Index'], inplace=True)

data_frame_copy
```
